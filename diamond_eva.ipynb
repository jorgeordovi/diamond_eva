```python
import numpy as np
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
import seaborn  as sns
```

```python
df=pd.read_csv('diamonds.csv')
```

```python
df.head
```

```python
bins = [0,500,1000,5000,10000,20000]
labels= ['Bajo','Medio-Bajo','Medio','Medio-Alto','Alto']
df['price_category'] = pd.cut(df['price'], bins=bins, labels=labels)
```

```python
labels
```

```python
df.head(200)
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['x'], df['price'], alpha=0.5, edgecolors='w', linewidth=0.5)
plt.title('Scatter Plot of Diamond Price vs. Size (Carat)')
plt.xlabel('Price')
plt.ylabel('Carat')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(df['price'], bins=60, edgecolor='k', alpha=0.7)
plt.title('Histogram of Diamond Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='clarity', y='depth', data=df)
plt.title('Boxplot of Diamond Depth by Clarity')
plt.xlabel('Clarity')
plt.ylabel('Depth')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(df['price_category'], bins=50, edgecolor='k', alpha=0.7)
plt.title('Distribution of Diamond Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['table'], df['depth'], alpha=0.5, edgecolors='w', linewidth=0.5)
plt.title('Scatter Plot of Diamond Price vs. Size (Carat)')
plt.xlabel('Price')
plt.ylabel('Carat')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.violinplot(x='cut', y='price', data=df)
plt.title('Violin Plot of Diamond Prices by Cut Category')
plt.xlabel('Cut')
plt.ylabel('Price')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.scatterplot(x='carat', y='depth', hue='color', data=df, alpha=0.6, edgecolor='k', linewidth=0.5)
plt.title('Scatter Plot of Diamond Carat vs. Depth, Colored by Color')
plt.xlabel('Carat')
plt.ylabel('Depth')
plt.legend(title='Color')
plt.grid(True)
plt.show()
```

```python
price_cut_counts = pd.crosstab(df['cut'], df['price_category'])

price_cut_counts.plot(kind='bar', stacked=True, figsize=(12, 6), colormap='viridis')
plt.title('Stacked Bar Plot of Diamond Price Categories by Cut')
plt.xlabel('Cut')
plt.ylabel('Count')
plt.legend(title='Price Category', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(axis='y')
```
